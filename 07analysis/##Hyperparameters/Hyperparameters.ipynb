```python
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
HYBRID_TRAINING_LOG = "./hybriddata/Hybridtest_results/"
BATCHSIZE = HYBRID_TRAINING_LOG +"BatchSize"
KERNEL_SIZE = HYBRID_TRAINING_LOG +"kernel_size"
MAX_LEN = HYBRID_TRAINING_LOG +"Max_len"
NUMFILTERS = HYBRID_TRAINING_LOG +"NumFilters"
```

```python
import ast
def parsing(file_name):
  info = {}
  for i in file_name.split("_"):
    if "lr" in i:
      info["lr"] = float(i[2:])
    elif "epoch" in i:
      info["epoch"] = int(i[i.rfind("epoch")+len("epoch"):])
    elif "threshold" in i:
      info["threshold"] = float(i[i.rfind("threshold")+len("threshold"):])
    elif "LEN" in i:
      info["MAX_LEN"] = int(i[i.rfind("LEN")+len("LEN"):])
    elif "SIZES" in i:
      info["kernel_sizes"] = ast.literal_eval(i[i.rfind("SIZES")+len("SIZES"):])
    elif "SIZE" in i:
      info["batch_size"] = int(i[i.rfind("SIZE")+4:])
    elif "FILTERS" in i:
      info["num_filters"] = int(i[i.rfind("FILTERS")+len("FILTERS"):-4])
  return info
```

```python
BATCHSIZE_info = [(filename, parsing(filename)) for filename in os.listdir(BATCHSIZE)]
KERNEL_SIZE_info = [(filename, parsing(filename)) for filename in os.listdir(KERNEL_SIZE)]
MAX_LEN_info = [(filename, parsing(filename)) for filename in os.listdir(MAX_LEN)]
NUMFILTERS_info = [(filename, parsing(filename)) for filename in os.listdir(NUMFILTERS)]
```

```python
BATCHSIZE_info = sorted(BATCHSIZE_info, key= lambda x : x[1]["batch_size"])
KERNEL_SIZE_info = sorted(KERNEL_SIZE_info, key= lambda x : len(x[1]["kernel_sizes"]))
MAX_LEN_info = sorted(MAX_LEN_info, key= lambda x : x[1]["MAX_LEN"])
NUMFILTERS_info = sorted(NUMFILTERS_info, key= lambda x : x[1]["num_filters"])
```

```python
def NUMFILTERS_ANALYSIS():
  # Metric lists
  Precision_list = []
  Recall_list = []
  F1_list = []
  Accuracy_list = []
  filename_list =[]
  info_list =[]
  for filename, info in NUMFILTERS_info:
    with open(os.path.join(NUMFILTERS, filename), "r") as f:
      data = f.read()
    data = data.split("\n")[:4]
    Precision = float(data[0][data[0].find(": ") + 2:].strip())
    Recall = float(data[1][data[1].find(": ") + 2:].strip())
    F1 = float(data[2][data[2].find(": ") + 2:].strip())
    Accuracy = float(data[3][data[3].find(": ") + 2:].strip())
    Precision_list.append(Precision)
    Recall_list.append(Recall)
    F1_list.append(F1)
    Accuracy_list.append(Accuracy)
    filename_list.append(filename)
    info_list.append(info)
  return {
    "Precision_list" : Precision_list,
    "Recall_list" : Recall_list,
    "F1_list" : F1_list,
    "Accuracy_list" : Accuracy_list,
    "filename_list" : filename_list,
    "info_list" : info_list,
  }
def BATCH_SIZE_ANALYSIS():
  # Metric lists
  Precision_list = []
  Recall_list = []
  F1_list = []
  Accuracy_list = []
  filename_list =[]
  info_list =[]
  for filename, info in BATCHSIZE_info:
    with open(os.path.join(BATCHSIZE, filename), "r") as f:
      data = f.read()
    data = data.split("\n")[:4]
    Precision = float(data[0][data[0].find(": ") + 2:].strip())
    Recall = float(data[1][data[1].find(": ") + 2:].strip())
    F1 = float(data[2][data[2].find(": ") + 2:].strip())
    Accuracy = float(data[3][data[3].find(": ") + 2:].strip())
    Precision_list.append(Precision)
    Recall_list.append(Recall)
    F1_list.append(F1)
    Accuracy_list.append(Accuracy)
    filename_list.append(filename)
    info_list.append(info)
  return {
    "Precision_list" : Precision_list,
    "Recall_list" : Recall_list,
    "F1_list" : F1_list,
    "Accuracy_list" : Accuracy_list,
    "filename_list" : filename_list,
    "info_list" : info_list,
  }
def MAX_LEN_ANALYSIS():
  # Metric lists
  Precision_list = []
  Recall_list = []
  F1_list = []
  Accuracy_list = []
  filename_list =[]
  info_list =[]
  for filename, info in MAX_LEN_info:
    with open(os.path.join(MAX_LEN, filename), "r") as f:
      data = f.read()
    data = data.split("\n")[:4]
    Precision = float(data[0][data[0].find(": ") + 2:].strip())
    Recall = float(data[1][data[1].find(": ") + 2:].strip())
    F1 = float(data[2][data[2].find(": ") + 2:].strip())
    Accuracy = float(data[3][data[3].find(": ") + 2:].strip())
    Precision_list.append(Precision)
    Recall_list.append(Recall)
    F1_list.append(F1)
    Accuracy_list.append(Accuracy)
    filename_list.append(filename)
    info_list.append(info)
  return {
    "Precision_list" : Precision_list,
    "Recall_list" : Recall_list,
    "F1_list" : F1_list,
    "Accuracy_list" : Accuracy_list,
    "filename_list" : filename_list,
    "info_list" : info_list,
  }
def KERNEL_SIZE_ANALYSIS():
  # Metric lists
  Precision_list = []
  Recall_list = []
  F1_list = []
  Accuracy_list = []
  filename_list =[]
  info_list =[]
  for filename, info in KERNEL_SIZE_info:
    with open(os.path.join(KERNEL_SIZE, filename), "r") as f:
      data = f.read()
    data = data.split("\n")[:4]
    Precision = float(data[0][data[0].find(": ") + 2:].strip())
    Recall = float(data[1][data[1].find(": ") + 2:].strip())
    F1 = float(data[2][data[2].find(": ") + 2:].strip())
    Accuracy = float(data[3][data[3].find(": ") + 2:].strip())
    Precision_list.append(Precision)
    Recall_list.append(Recall)
    F1_list.append(F1)
    Accuracy_list.append(Accuracy)
    filename_list.append(filename)
    info_list.append(info)
  return {
    "Precision_list" : Precision_list,
    "Recall_list" : Recall_list,
    "F1_list" : F1_list,
    "Accuracy_list" : Accuracy_list,
    "filename_list" : filename_list,
    "info_list" : info_list,
  }


#########
def get_plot(analysis_data, Target, label):
  Precision_list = analysis_data["Precision_list"]
  Recall_list = analysis_data["Recall_list"]
  F1_list = analysis_data["F1_list"]
  Accuracy_list = analysis_data["Accuracy_list"]
  info_list = analysis_data["info_list"]
  target_info_list = [info[label] for info in info_list]
  # 각 메트릭의 시계열 플롯
  plt.figure(figsize=(15, 8))
  plt.subplot(2, 2, 1)
  plt.plot(Precision_list, label='Precision', marker='o')
  plt.title('Precision By %s'%Target)
  plt.xlabel('Run')
  plt.ylabel('Precision')
  plt.xticks(np.arange(0,len(target_info_list)), labels=target_info_list,rotation=45)
  plt.grid()
  # plt.ylim(0, 1)

  plt.subplot(2, 2, 2)
  plt.plot(Recall_list, label='Recall', marker='o')
  plt.title('Recall By %s'%Target)
  plt.xlabel('Run')
  plt.ylabel('Recall')
  plt.xticks(np.arange(0,len(target_info_list)), labels=target_info_list,rotation=45)
  plt.grid()
  # plt.ylim(0, 1)

  plt.subplot(2, 2, 3)
  plt.plot(F1_list, label='F1 Score', marker='o')
  plt.title('F1 Score By %s'%Target)
  plt.xlabel('Run')
  plt.ylabel('F1 Score')
  plt.xticks(np.arange(0,len(target_info_list)), labels=target_info_list,rotation=45)
  plt.grid()
  # plt.ylim(0, 1)

  plt.subplot(2, 2, 4)
  plt.plot(Accuracy_list, label='Accuracy', marker='o')
  plt.title('Accuracy By %s'%Target)
  plt.xlabel('Run')
  plt.ylabel('Accuracy')
  plt.xticks(np.arange(0,len(target_info_list)), labels=target_info_list,rotation=45)
  plt.grid()
  # plt.ylim(0, 1)

  plt.tight_layout()
  plt.show()
```

```python
batch_size_analysis = BATCH_SIZE_ANALYSIS()
max_len_analysis=MAX_LEN_ANALYSIS()
numfilters_analysis=NUMFILTERS_ANALYSIS()
kernel_size_analysis=KERNEL_SIZE_ANALYSIS()
  # plt.savefig(os.path.join(plot_directory, MODEL_NAME+'metrics_over_runs.png'))
```

```python
get_plot(batch_size_analysis, "Batch Size", "batch_size")
get_plot(max_len_analysis, "Max Len", "MAX_LEN")
get_plot(numfilters_analysis, "Num Filters","num_filters")
get_plot(kernel_size_analysis, "Kernel", "kernel_sizes")
```
