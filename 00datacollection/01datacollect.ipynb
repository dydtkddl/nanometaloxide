## Springer에서 각 물질들에 대한 논문을 논문 제목으로써 검색

#### API 쿼리로 쓸 쿼리 스트링을 만들기

```python
import requests
APIKEYS = ["52035a92a974ae59d4bc929c6a8ae9de",
           "54ef0223aa5d90bd0febe21634994901",
           "53bca5462c6d35bb58e8247196d638e6",
           "119ecae1ca9127f1d96a6ebf9b536805",
           "1eea559f9c6503ce0722a8a3c3d958e7" ,
           "d19ac693c461be608b928cca2b18ab0f" ]
url = "http://api.springernature.com/meta/v2/json?q=(%sAND title:'nanoparticle')&api_key=%s&p=%s&s=%s"
query_dic = {}
for key in oxide_synonyms.keys():
  query = 'title:"'+key+'" OR '
  for i, synonym in  enumerate(oxide_synonyms[key]):
    length = len(oxide_synonyms[key])
    if length-1 == i:
      synonym = 'title:"' + synonym + '" '
    else:
      synonym = 'title:"' + synonym + '" OR '
    query += synonym
  query_dic[key] = query
  '''{'SiO2': 'title:"Silicon Dioxide" OR title:"Silica" OR title:"Quartz" OR title:"Cristobalite" OR title:"Tridymite"',
 'ZnO': 'title:"Zinc Oxide" OR title:"ZnO" OR title:"Zincite"',
 'TiO2': 'title:"Titanium Dioxide" OR title:"Titania" OR title:"Rutile" OR title:"Anatase" OR title:"Brookite"',
 'Al2O3': 'title:"Aluminum Oxide" OR title:"Alumina" OR title:"Corundum" OR title:"Alundum"',
 'Fe3O4': 'title:"Magnetite" OR title:"Iron(II,III) Oxide" OR title:"Fe3O4" OR title:"Ferrous Ferric Oxide"',
 'CuO': 'title:"Copper(II) Oxide" OR title:"CuO" OR title:"Tenorite"',
 'Fe2O3': 'title:"Iron(III) Oxide" OR title:"Ferric Oxide" OR title:"Hematite" OR title:"Maghemite"',
 'Cu2O': 'title:"Copper(I) Oxide" OR title:"Cuprous Oxide" OR title:"Cu2O" OR title:"Red Copper Oxide"',
 'CeO2': 'title:"Cerium Oxide" OR title:"Ceria" OR title:"CeO2" OR title:"Ceric Oxide"',
 'Mn2O3': 'title:"Manganese(III) Oxide" OR title:"Mn2O3" OR title:"Bixbyite"',
 'In2O3': 'title:"Indium Oxide" OR title:"In2O3" OR title:"Indium Sesquioxide"',
 'Fe3O4/TiO2': 'title:"Magnetite/Titania Composite" OR title:"Fe3O4/TiO2 Composite"',
 'Au/TMC/Fe3O4': 'title:"Gold/TMC/Magnetite Composite" OR title:"Au/TMC/Fe3O4 Composite"',
 'SPION': 'title:"Superparamagnetic Iron Oxide Nanoparticles" OR title:"SPIONs"',
 'MnO': 'title:"Manganese(II) Oxide" OR title:"MnO" OR title:"Manganosite"',
 'CdO': 'title:"Cadmium Oxide" OR title:"CdO" OR title:"Cadmium Monoxide"',
 'Graphene oxide': 'title:"Graphene Oxide" OR title:"GO"',
 'Graphene': 'title:"Graphene" OR title:"Single-Layer Graphite" OR title:"Monolayer Graphene"',
 'Co3O4': 'title:"Cobalt(II,III) Oxide" OR title:"Co3O4" OR title:"Cobaltosic Oxide"',
 'CuZnFe2O4': 'title:"Copper Zinc Ferrite" OR title:"CuZnFe2O4"',
 'HfO2': 'title:"Hafnium Oxide" OR title:"Hafnia" OR title:"HfO2" OR title:"Hafnium Dioxide"',
 'CoFe2O4': 'title:"Cobalt Ferrite" OR title:"CoFe2O4" OR title:"Cobalt Iron Oxide"'}'''
```

#### requests로 title정보 수집

```python
from tqdm import tqdm
import json
data2 = {}
for key,value in tqdm(query_dic.items()):
  response = requests.get(url%(value, APIKEYS[0],100, 1))
  if response.status_code == 200:
    json_response = json.loads(response.text)
    total = json_response["result"][0]["total"]
    records = json_response["records"]
    print(len(records))
    data2[key] = {"total" : total, "records" : records}
```

## title keyword url 정보 수집

```python
extracted_data = {}
for key, value in data2.items():
  total = value["total"]
  records = value["records"]
  sub_list = []
  for record in records:
    subdic = {}
    keyword = record.get("keyword", None)
    url = record.get("url", None)
    title = record.get("title", None)
    subdic["keyword"] = keyword
    subdic["url"] = url
    subdic["title"] = title
    sub_list.append(subdic)
  extracted_data[key] = sub_list
```

## 엑셀로 물질별로 100개씩 저장

```python
for key, value in extracted_data.items():
  ke = []
  ti = []
  for v in value:
    ke.append(v["keyword"])
    ti.append(v["title"])
  df = pd.DataFrame({"keyword": ke, "title":ti})
  if (len(df)>0):
    df.to_excel("%s_nanoparticle_data.xlsx"%key)
```
